```python
df['FirstTransactionDate'] = df.groupby('trainee_ID')['DATE(tran_date)'].transform('min')
```

```python
df['YearOfFt']=df['FirstTransactionDate'].dt.year
```

```python
df['MonthOfFt']=df['FirstTransactionDate'].dt.month_name()
```

```python
df['Diff_in_months']=(df['DATE(tran_date)'].dt.year - df['FirstTransactionDate'].dt.year)*12 + (df['DATE(tran_date)'].dt.month - df['FirstTransactionDate'].dt.month)
```
